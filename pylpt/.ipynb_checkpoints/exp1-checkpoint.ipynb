```python
from ltp import LTP
```

```python
ltp = LTP()
```

```python
# 分句
sents = ltp.sent_split(["他叫汤姆去拿外衣。", "汤姆生病了。他去了医院。"])
print(sents)
```

```python
segment, _ = ltp.seg(["他叫汤姆去拿外衣。"])
print(segment)

# 对于已经分词的数据
segment, hidden = ltp.seg(["他/叫/汤姆/去/拿/外衣/。".split('/')], is_preseged=True)
print(segment)
print(hidden)
```

# 词性
"他叫汤姆去拿外衣"
他：代词 r
叫：动词 v
汤姆：人名 nh
去：动司 v
拿：动词 v
外衣：一般名词 n
。：标点符合 wp

```python
# 每个词输出词性
seg, hidden = ltp.seg(["他叫汤姆去拿外衣。"])
pos = ltp.pos(hidden)
print(seg)
print(pos)
```

# 命名实体识别
识别文本中具有特殊意义的实体，Named Entity Recognition

```python
# 命名实体识别
seg, hidden = ltp.seg(["他叫汤姆去拿外衣。"])
ner = ltp.ner(hidden)
print(ner)

tag, start, end = ner[0][0]
print(tag,":", "".join(seg[0][start:end + 1]))
# Nh : 汤姆

seg,hidden = ltp.seg(["我下午三点半需要去趟中日友好医院，麻烦帮我预约张三医生"])
ner = ltp.ner(hidden)
print(seg)
print(ner)
```

```python
# 语义角色标注
seg,hidden = ltp.seg(["我下午三点半需要去趟中日友好医院，麻烦帮我预约张三医生"])
srl = ltp.srl(hidden)
print(srl)
```
